```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2016-10-01&endtime=2016-10-02'
req = requests.get(url).json()
```

```python
req.keys()
```

```python
type(req['features'])  # >> List  
```

```python
len(req['features']) # 309
```

```python
type(req['features'][0])  # >> dict  
```

```python
lat=[]
lon=[]
mag=[]
title =[]
place = []
for i in req['features']:

    lat.append(i['geometry']['coordinates'][0])
    lon.append(i['geometry']['coordinates'][1])
    mag.append(i['properties']['mag'] )
    title.append(i['properties']['title'] )
    place.append(i['properties']['place'] )
```

```python
print(len(lat))
print(len(lon))
print(len(mag))
print(len(title))
print(len(place))
```

```python
import pandas as pd
df = pd.DataFrame({'Title' : title , 'Place' : place , 'Magnitude' : mag , 
                   'Latitude' : lat ,'Longitude' : lon })
```
