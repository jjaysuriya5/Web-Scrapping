```python
from selenium import webdriver
import os 
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
```

```python
path ='C:\\Users\\Administrator\\Desktop\\Python\\Data Scraping\\Day 1'
os.chdir(path)
browser = webdriver.Firefox()
url = 'https://www.oyorooms.com/search?location=Coimbatore%2C%20Tamil%20Nadu%2C%20India&city=Coimbatore&searchType=city&coupon=&checkin=24%2F09%2F2019&checkout=25%2F09%2F2019&roomConfig%5B%5D=2&showSearchElements=false&country=india&guests=2&rooms=1&filters%5Bcity_id%5D=82'
browser.get(url)
html = browser.page_source
soap = BeautifulSoup(html)
```

```python
hotel_name = [i.text for i in soap.find_all('h3'  , class_ ='listingHotelDescription__hotelName d-textEllipsis' )]
address = [ i.text for i in soap.find_all('span' , itemprop = 'streetAddress')]
ratings = [ i['content'] for i in soap.find_all('meta',itemprop="ratingValue") ]
ratings_count = [ i['content'] for i in soap.find_all('meta',itemprop="reviewCount") ]
discounted_price = [ i.text for i in soap.find_all('span', class_='listingPrice__finalPrice')]
orginal_price = [ i.text for i in soap.find_all('span', class_='listingPrice__slashedPrice d-body-lg')]
discount_pct = [ i.text for i in soap.find_all('span', class_='listingPrice__percentage')]
```

```python
table = pd.DataFrame()
table['Hotel_Name'] = hotel_name
table['Address'] = address
table['Ratings'] = ratings
table['No_of_Users_Rated'] = ratings_count
table['Price'] = discounted_price
table['Orginal_Price'] = orginal_price
table['Discount_Percentage'] = discount_pct

table['Price'] = table['Price'].apply(lambda x : x.replace ('₹' , '')).astype(int)
table['Orginal_Price'] = table['Orginal_Price'].apply(lambda x : x.replace ('₹' , '')).astype(int)
table['Discount_Percentage'] = table['Discount_Percentage'].apply(lambda x : x.replace ('% off' , '')).astype(float)
table['No_of_Users_Rated'] = table['No_of_Users_Rated'].apply(int)
table['Ratings'] = table['Ratings'].apply(float)
```
